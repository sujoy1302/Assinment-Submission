# Line Graph

```python

import matplotlib.pyplot as plt
# list of x-axis values
x=[5,2,6,4,9]

#list of y-axis values...
y=[10,4,12,8,18]

#function to plot with the arguments
plt.plot(x,y) #plot is a function to plot line chart...

# function to show the plotting 
plt.show()
```

# Bar Graph

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.bar(x,y)
# to have a title for ur graph...
plt.title("My Bar Graph")
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
values=[10,100.1,100.2,120,120.5]
plt.hist(values)
plt.show()
```

# scatter

```python
import matplotlib.pyplot as plt
x=[1,4,13,15,56,155,122,55]
y=[13,55,31,1,66,5,100,15]
plt.scatter(x,y)
plt.title("Scatter Graph")
# labeling the axis 
plt.xlabel("Time (hr)")
plt.ylabel("Kilometers (km)")
plt.show()
```
